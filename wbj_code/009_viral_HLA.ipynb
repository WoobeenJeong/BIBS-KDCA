```python
def short_view(directory):
    all_dfs = []
    filenames = os.listdir(directory) 
    if filenames:
        filename = filenames[10]
        if filename.startswith('COV') and filename.endswith('.csv') or filename.endswith('.txt'):
            df = pd.read_csv(directory + filename, sep=',')
            df['Directory'] = directory.split('/')[-2]
            df.insert(0, 'Directory', df.pop('Directory'))
            all_dfs.append(df.head(n=3))
            
    result = pd.concat(all_dfs)
    display(result)
```

```python
### cyt_list의 경로에 있는 모든 파일에 대해 
# Sample.ID.	alpha.2.Macroglobulin..pg.ml.	beta.2.Microglobulin..pg.ml.	Fetuin.A.AHSG..pg.ml.	MBL..pg.ml.	Protein.S.PROS1..pg.ml.	Serpin.A4.Kallistatin..pg.ml.	CD14..pg.ml.	Component.C2..pg.ml.	CXCL4.PF4..pg.ml.	LBP..pg.ml.	LRG1..pg.ml.	Properdin..pg.ml.	Serpin.C1..pg.ml.	Serpin.A10.ZPI..pg.ml.	Aldehyd.Dehydrogenase.1.A1..pg.ml.	BCMA.TNFRSF17..pg.ml.	CCL14..pg.ml.	CCL18.PARC..pg.ml.	Coagulation.Factor.XIV.Protein.C..pg.ml.	Collagen.I.alpha.1..pg.ml.	Complement.Factor.D.Adipsin..pg.ml.	DPPIV.CD26..pg.ml.	Fibroblast.Activation.Protein.alpha.FAP..pg.ml.	Galectin.3..pg.ml.	Galectin.3BP.MAC.2BP..pg.ml.	IGFBP.2..pg.ml.	Lumican..pg.ml.	Myeloperoxidase.MPO..pg.ml.	VAP.1.AOC3..pg.ml.	CA15.3.MUC.1..Unit.	CCL5.RANTES..pg.ml.	Complement.component.C9..pg.ml.	EN.RAGE.S100A12..pg.ml.	M.CSF.R.CD115..pg.ml.	MMP.2..pg.ml.	MMP.9..pg.ml.	Myoglobin..pg.ml.	Lipocalin.2.NGAL..pg.ml.	TIMP.1..pg.ml.	CCL19.MIP.3.beta..pg.ml.	CCL2.MCP.1..pg.ml.	CCL3.MIP.1.alpha..pg.ml.	CCL4.MIP.1.beta..pg.ml.	CD23.Fc.epsilon.RII..pg.ml.	CD31.PECAM.1..pg.ml.	CX3CL1.Fractalkine..pg.ml.	CXCL9.MIG..pg.ml.	ErbB3.Her3..pg.ml.	Flt.3.Ligand.FLT3L..pg.ml.	Granzyme.B..pg.ml.	HE4.WFDC2..pg.ml.	IFN.gamma..pg.ml.	IGFBP.4..pg.ml.	IL.1.beta..pg.ml.	IL.10..pg.ml.	IL.12.p40..pg.ml.	IL.6..pg.ml.	IL.8..pg.ml.	Kallikrein.6.Neurosin..pg.ml.	LIGHT.TNFSF14..pg.ml.	MMP.12..pg.ml.	MMP.13..pg.ml.	MMP.3..pg.ml.	MMP.7..pg.ml.	MMP.8..pg.ml.	TGF.alpha..pg.ml.	TNF.alpha..pg.ml.	TREM.1..pg.ml.	TSLP..pg.ml.	VEGF..pg.ml.	Alpha.1.Microglobulin..pg.ml.	Angiopoietin.2..pg.ml.	BAFF..pg.ml.	BMP.10..pg.ml.	Contactin.1..pg.ml.	CXCL11.I.TAC..pg.ml.	CXCL13.BCA.1..pg.ml.	CXCL5.ENA.78..pg.ml.	Endocan.ESAM.1..pg.ml.	ErbB2.Her2..pg.ml.	FGF.basic.bFGF..pg.ml.	Galectin.9..pg.ml.	G.CSF..pg.ml.	GM.CSF..pg.ml.	ICAM.1.CD54..pg.ml.	IL.1.alpha.IL.1F1..pg.ml.	IL.12.p70..pg.ml.	IL.17.IL.17A..pg.ml.	IL.1ra.IL.1F3..pg.ml.	IL.2..pg.ml.	IL.31..pg.ml.	IL.4..pg.ml.	IL.5..pg.ml.	Leptin.OB..pg.ml.	MAdCAM.1..pg.ml.	NT.4..pg.ml.	SCF.c.kit.Ligand...pg.ml.	Thrombopoietin.Tpo..pg.ml.	u.Plasminogen.Activator..uPA...pg.ml.	VCAM.1.CD106..pg.ml.	CCL22.MDC..pg.ml.	CCL23.MPIF.1..pg.ml.	CCL26.Eotaxin.3..pg.ml.	CD117.c.kit..pg.ml.	CD163..pg.ml.	CEACAM.1.CD66a..pg.ml.	Coagulation.Factor.III.Tissue.Factor..pg.ml.	Follistatin.related.Gene.Protein.FLRG..pg.ml.	Furin..pg.ml.	IFN.beta..pg.ml.	IGFBP.1..pg.ml.	IL.11..pg.ml.	IL.28B.IFN.lamgda.3..pg.ml.	IL.4R.alpha..pg.ml.	IL.6R.alpha..pg.ml.	MCAM.CD146..pg.ml.	Oncostatin.M.OSM...pg.ml.	Osteoactivin.GPNMB..pg.ml.	Reg3A..pg.ml.	Resistin..pg.ml.	S100A9..pg.ml.	SCGF.CLEC11a..pg.ml.	Serpin.B3.SCCA1..pg.ml.	TACI.TNFRSF13B..pg.ml.	Thrombomodulin.BDCA.3..pg.ml.	Thrombospondin.2..pg.ml.	TRAIL.R2.TNFRSF10B..pg.ml.	TRNACE.TNFSF11.RANK.L..pg.ml.	ADAMTS13..pg.ml.	APRIL.TNFSF13..pg.ml.	CA125.MUC16..Units.	Chitinase.3.like.1..pg.ml.	CXCL1.GRO.alpha..pg.ml.	CXCL10.IP.10..pg.ml.	Dkk.1..pg.ml.	EGF..pg.ml.	ENPP.2.Autotaxin..pg.ml.	GDF.15..pg.ml.	Growth.Hormone..pg.ml.	HGF..pg.ml.	IFN.alpha..pg.ml.	IL.13..pg.ml.	IL.18.IL.1F4...pg.ml.	IL.23..pg.ml.	IL.33..pg.ml.	M.CSF..pg.ml.	MICA..pg.ml.	MIF..pg.ml.	Neuregulin.1.beta.NRG1.beta.1..pg.ml.	Pentraxin.3.TSG.14..pg.ml.	SP.D..pg.ml.	ST2.IL.33R..pg.ml.	uPAR..pg.ml.	vWF.A2..pg.ml.	Angiopoietin.1..pg.ml.	Angiopoietin.like.Protein.3.ANGPTL3..pg.ml.	CCL11.Eotaxin..pg.ml.	CCL24.Eotaxin.2.MPIF.2..pg.ml.	CD40.Ligand.TNFSF5..pg.ml.	Complement.Componet.C5a..pg.ml.	CXCL2.Gro.Beta.MIP.2..pg.ml.	CXCL6.GCP.2..pg.ml.	FABP4.A.FABP..pg.ml.	Fas.Ligand.TNFSF6..pg.ml.	Granzyme.A..pg.ml.	IL.15..pg.ml.	IL.17E.IL.25..pg.ml.	IL.3..pg.ml.	IL.36.beta.IL.1F8..pg.ml.	IL.7..pg.ml.	LIF..pg.ml.	L.Selectin.CD62L..pg.ml.	MIA..pg.ml.	Nectin.4...pg.ml.	Nephrin..pg.ml.	Osteopontin.OPN..pg.ml.	PDGF.AA..pg.ml.	PD.L1.B7.H1..pg.ml.	Prolactin..pg.ml.	Proprotein.convertase.9.PCSK9..pg.ml.	P.Selectin.CD62P..pg.ml.	Syndecan.1.CD138..pg.ml.	TFF3..pg.ml.	TRAIL.TNFSF10..pg.ml.	VEGFR1.Flt.1..pg.ml.	TGF.beta.1..pg.ml.	TGF.beta.2..pg.ml.	BDNF..pg.ml.	PDFG.DD..pg.ml.	Complement.C1q..ng.ml.	SDF.1..pg.ml.
# CCO-001-1	411539661.8 	626906.9 	171026177.5	1448933.7 	3532220.8	11603412.1	850775.5	17326498.7	808855.5	12877186.4	5841211.6	7484792.8	1737816.4	3123555.2	27173.3	7992.3	16683.4	9122.3	44607.5 	2655	1792873.4	96783.8	7258.2	5230.5	853728	94706.6	494400.7	59423.5	84398.2	5.8	14816.7	106646560.2	37050.2	113773.6	91854.8	46868.1	2186.1	25639	41856.3	59.5	350.8	244.4 	318.4 	1163.6 	40031.3	2976.5 	687.3 	8149.4	53.6	30.4	3445.9	6.2 	75.3 	<1.34	4.8 	375.5 	5.9 	9.9 	1247.4	0.5 	<38.97	700.6 	18520.3	3081.7 	2038	4.1 	5.1 	77.6 	4.1 	36.3 	51600493	292.3 	372.1	60.1 	2284.8	89.6 	54.1 	207.3 	318.6	1196.3 	13.2 	27781.5	71.6 	9.6 	139665.3 	6.5 	31.7 	10.0 	838.3	13.6 	<69.26	47.4 	5.3 	2782.2 	122383.2	39.0 	25.3 	130.0 	335.4	1528003.4	139.6	188.6	33.5 	4133.7 	283797.7	42035.3 	32.7	9184.9	2907.5	16.0 	35615.9 	2588.7 	237.8 	24.8	28539	25967.2	1163.7	7929.3	6701.3	6523.3	435.5	5438.9	720.5 	29.9 	3718.9	14598.9	30.1 	9.3 	811629.7	287.5 	67.9 	10476.9	232.1 	127.5	629	43.5 	23827.3	535.6	457.0 	79.1	11.7 	358.6 	274	205.3 	0.7 	506.5 	57.8 	10221.2	<7.22	966.8 	4404.2	10444.4	237.7 	8185.9 	8180.7	26597.8	307.9	235.9	1950.9	151077.9	709.3	165.9	4430.5	49.2	138	25.7	1975.9	150.2 	49.2	31.5	57.4 	296591.8	455.4 	413.9	889.6	11869.4	459	56.8	20037.7	70972.4	24774.7	2362.2	1010	202.6 	220.2	16660.9 	<13.81	1940.1	250.7	43336.2 	2185.8 
### 모든 txt 파일에서 첫 줄 column 이름만 가져오기

### 만약 경로가 dir_CYT_01이면 '\t'을, dir_CYT_02이면 ','를 구분자로 사용

def replace_special_chars(col_name):
    col_name = re.sub(r'[ \-\(\),/]', '.', col_name)
    col_name = re.sub(r'\.{2,}', '.', col_name)
    return col_name

def treat_df(df):
    df.columns = [replace_special_chars(col) for col in df.columns]
    df = df.rename(columns={'Sample.ID.': 'ID'})
    ### 만약 ID에 CNC 또는 CNO가 없다면
    if not any(df['ID'].str.contains('CNC|CNO')):
        df['PNTTM'] = df['ID'].str[-1]
        df['PNTTM'] = df['PNTTM'].replace('0', '1')
        df['ID'] = df['ID'].str[:-2]
    else:
        df['PNTTM'] = '1'
    df.columns = [col[1:] if col.startswith('.') else col for col in df.columns]
    return df

def get_cytokine(directory):
    all_dfs = []
    filenames = os.listdir(directory)
    for filename in filenames:
        if filename.startswith('COV') and filename.endswith('.txt'):
            if directory == dir_CYT_01:
                df = pd.read_csv(directory + filename, sep='\t')
                df = treat_df(df)
                df['ID'] = 'COV-' + df['ID']
            else:
                df = pd.read_csv(directory + filename, sep=',')
                df = treat_df(df)
                if df['ID'].str.contains('CNC|CNO').any():
                    df['ID'] = df['ID'].str[:-2]

            all_dfs.append(df)
    result = pd.concat(all_dfs)
    return result

cyt_01 = get_cytokine(dir_CYT_01)
cyt_02 = get_cytokine(dir_CYT_02)
cyt_all = pd.concat([cyt_01, cyt_02])

print(cyt_all.columns)
```

```python
### cyt_adapt에서 맨 뒤의 7자리를 제외하고 n글자 이상 동일한 문자열이 있는 column 확인



def get_similar_cols(df, length):
    similar_cols = []
    
    def get_substrings(s):
        if len(s) < length:
            return []
        return [s[i:i+length] for i in range(len(s) - length + 1)]
    
    for col in df.columns:
        if len(col) > length:
            col_substrings = set(get_substrings(col[:-7]))
            for col2 in df.columns:
                if col != col2 and len(col2) > length:
                    col2_substrings = set(get_substrings(col2[:-7]))
                    if col_substrings.intersection(col2_substrings):
                        similar_cols.append(col)
                        break
    return similar_cols

similar_cols = get_similar_cols(cyt_adapt,15)
for i in similar_cols:
    print(i, sep='\n')
print(len(similar_cols))
        
for i in similar_cols:
    if not i.endswith('pg.ml.'):
        print(i)
```
